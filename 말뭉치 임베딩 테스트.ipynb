```python
# 오류 주석 딕셔너리
error_pattern_dict = {
    "OM": "누락",    # 어휘나 문법 요소가 빠진 경우
    "ADD": "첨가",   # 불필요한 요소가 덧붙여진 경우
    "REP": "대치",   # 다른 표현으로 바뀐 경우
    "MIF": "오형태"  # 형태가 틀린 경우 (예: 어미나 어간의 오류)
}

# 오류 층위 딕셔너리
error_annotation = {
    # 발음 오류
    "PP": "음소",                    # 음소 오류
    "PS": "음절",                   # 음절 오류
    "PC": "음운규칙",               # 음운 규칙 오류
    "PN": "원어식 발음",            # 원어식 발음 (임시 기호)
    "PA": "중간 발음(변이음 포함)",   # 변이음 포함 중간 발음 (임시 기호)

    # 형태 오류
    "MCP": "단어 형성[합성법]",       # 합성법 오류
    "MDV": "단어 형성[파생법]",       # 파생법 오류
    "MDC": "굴절[곡용]",             # 곡용 오류
    "MCJ": "굴절[활용]",             # 활용 오류
    "POS": "품사",                  # 품사 오류

    # 통사 오류
    "SH": "높임",                   # 높임 표현 오류

    # 담화 오류
    "ST": "시제",                   # 시제 표현 오류
    "SC": "사동",                   # 사동 표현 오류
    "SP": "피동",                   # 피동 표현 오류
    "SN": "부정",                   # 부정 표현 오류
    "WO": "어순",                   # 어순 오류
    "DR": "지시",                   # 지시 표현 오류 (deixis)
    "DC": "접속",                   # 접속 표현 오류
    "DM": "담화표지",               # 담화표지 오류
    "DS": "구어/문어 오류"           # 말글체 불일치 오류
}

def parser(key, df):
    """
    문장 그룹별 오류 정보를 구조화하여 반환하는 함수

    Parameters:
        key (tuple): ('표본 번호', '문장') 그룹 키
        df (DataFrame): 해당 문장의 형태소 분석 정보

    Returns:
        dict: 구조화된 문장 오류 정보
    """
    wrong_sentence = key[1]
    wrong_morphs = []
    wrong_pos = []
    corrections = []
    idx = 0
    for _, row in df.iterrows():
        if row["원 형태소"] != '0' or row["형태 주석"] != '0':
            wrong_morphs.append(row["원 형태소"])
            wrong_pos.append(row["형태 주석"])
            idx = idx + 1

        error_code = row["오류 양상"]
        error_annotation_code = row["오류 층위"]
        if error_code != '0':  # 오류가 명시된 경우만 처리
            corrections.append({
                "index": idx,  # wrong_morphs 기준 위치
                "action": error_pattern_dict.get(error_code, "기타"),
                "annotation": error_annotation.get(error_annotation_code, "기타"),
                "morph": str(row["교정 형태소"]),
                **({"pos": row["형태 주석"]} if row["형태 주석"] != '0' else {})
            })

    return {
        "wrong_sentence": wrong_sentence,
        "wrong_morphs": wrong_morphs,
        "wrong_pos": wrong_pos,
        "corrections": corrections,
        "metadata": {
            "학습 환경": row["학습 환경"],
            "자료 유형": row["자료 유형"],
            "작문 유형": row["작문 유형"],
            "장르": row["장르"],
            "주제": row["주제"],
            "어절": row["어절"],
            "수집 대상": row["수집 대상"],
            "학습자 유형": row["학습자 유형"],
            "한국어 등급": row["한국어 등급"],
            "국적": row["국적"]
        }
    }
```

```python
import json

idx = 0

for key, group in tqdm(groupped, total=len(groupped), desc="말뭉치 저장 중"):
    result = parser(key, group)
    collection.add(
        ids=str(idx),
        documents=key[1],
        embeddings=model.encode(key[1]).tolist(),
        metadatas={
            "wrong_morphs": '|'.join(result['wrong_morphs']),
            "wrong_pos": '|'.join(result['wrong_pos']),
            "corrections": json.dumps(result['corrections']),
            **(result['metadata'])
        }
    )
    idx = idx + 1
```

```python
# 질문 / 벡터화
user_query = "언니 두 명 곌혼 있습니다"

query_embedding = model.encode(user_query)
```

```python
# 검색 (Top-2 결과 반환)
results = collection.query(
    query_embeddings=query_embedding.tolist(),
    n_results=2
)

# 검색 결과 출력
for doc, score, meta in zip(results['documents'][0], results['distances'][0], results['metadatas'][0]):
    print(f"문서: {doc}")
    print(f"유사도 거리: {score:.4f}")
    print(f"메타데이터: {meta}")
    print(json.loads(meta['corrections']))
    print("-" * 40)
```
